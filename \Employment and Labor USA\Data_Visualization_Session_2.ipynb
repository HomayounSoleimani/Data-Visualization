```
import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
import pandas as pd

import dash
import dash_core_components as dcc
import dash_html_components as html
```

```
#https://www.kaggle.com/maransk/labor-percapita-income-us-county/metadata
```

```
df_sample = pd.read_csv('https://raw.githubusercontent.com/HomayounSoleimani/Data-Visualization/master/Employment%20and%20Labor%20USA/ACS_17_5YR_DP03_with_ann_labour_economic_data_only.csv')
df_grouped = df_sample.groupby(df_sample.State).sum()
```

```
states=df_sample.State.unique()

fig = go.Figure(data=[
    go.Bar(name='Employee_pop', x=states, y=df_grouped.Employee_pop),
    go.Bar(name='labor_pop', x=states, y=df_grouped.labor_pop)
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
```

```
df_sample_r = df_sample[df_sample['State'] == 'California']
df_sample_r = df_sample_r.dropna()

values = df_sample_r['labor_pop'].tolist()
fips = df_sample_r['FIPS'].tolist()

colorscale = [
    'rgb(250, 163, 7)',
    'rgb(232, 93, 4)',
    'rgb(208, 0, 0)',
    'rgb(106, 4, 15)',
    'rgb(3, 7, 30)'
]

fig = ff.create_choropleth(
    fips=fips, values=values, scope=['California'],
    binning_endpoints=[10527, 30231, 56261, 90956], colorscale=colorscale, # 10527: 25%, 30231 50%, 56261 mean, 90956 75%
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5}, round_legend_values=True,
    legend_title='Labour Population by State', title='California Labour Population'
)
fig.layout.template = None
fig.show()
```

```
fig = px.pie(df_sample, values='labor_pop_percent', names='State', title='Labor Population Percentage over USA States')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```
fig = px.scatter(df_sample, x="labor_pop", y="Employee_pop", color="State",
                 size='Employee_pop', hover_data=['labor_pop'])
fig.show()
```
